```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('hirable.csv')
```

```python
df
```

```python
df = df.drop([
    "sl_no",
    "ssc_p",
    "ssc_b",
    "hsc_p",
    "hsc_b",
    "hsc_s",
    "specialisation",
    "salary",
    "degree_t"
], axis=1)
df = df.rename(columns = {'degree_p': 'bsc', 'mba_p': 'msc'})
df['gender'] = df.gender.replace(['M', 'F'], [1, 2])
df['workex'] = df.workex.replace(['Yes', 'No'], [1, 0])
df['status'] = df.status.replace(['Placed', 'Not Placed'], [1, 0])
```

```python
df
```

```python
def downscale(score):
    return score/10/2
degrees = ['bsc', 'msc']
for col in degrees:
    df[col] = downscale(df[col])
```

```python
X = df.drop(['status'], axis=1)
y = df.status
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X,y,train_size=0.8,random_state=1)
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier(n_estimators=100)
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)

# Report and Accuracy Score
from sklearn import metrics
from sklearn.metrics import classification_report
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
print("Classification Report RF:\n",classification_report(y_test,y_pred))
```

```python
# Model testing on new data
# [[gender, bsc, workex, etest_p, msc]]

# Sample 1
sample = np.array([[0, 2.9, 1, 78.50, 3.7]])
model.predict(sample)

# Sample 2
sample = np.array([[0, 2.9, 1, 78.50, 3.7]])
model.predict(sample)
```

```python
# Saving model
import pickle
pickle.dump(model, open('hireable.pkl', 'wb'))
```

```python
loaded_model = pickle.load(open('hireable.pkl', 'rb'))
result = loaded_model.score(X_test, y_test)
print(result)
```
